```python
import os

from groq import Groq

client = Groq(
    api_key=os.getenv("GROQ_API_KEY"),
)
```

```python
mensagem = [
    {
        "role": "user",
        "content": "Explique a diferença entre calvinista e arminiando. Responda em portugês"
    }
]

response = client.chat.completions.create(
    messages=mensagem,
    model="llama-3.1-8b-instant"
)

print(response.choices[0].message.content)
```

### Stream

```python
mensagem = [
    {
        "role": "user",
        "content": "Explique a diferença entre calvinista e arminiando. Responda em portugês"
    }
]

stream = client.chat.completions.create(
    messages=mensagem,
    model="llama-3.1-8b-instant",
    stream=True
)

for chunk in stream:
    print(chunk.choices[0].delta.content, end="", flush=True)
```

### Transcrição de áudio

```python
import textwrap

def format_text(response):
    text = response.text
    text_formatted = textwrap.fill(text, width=100)
    print(text_formatted)
```

```python
from groq import Groq

client = Groq()

arquivo = 'audios\SOMOS INGRATOS_!_! _ Juliana Cruz _ Lucas 17_11-19 #devocional #elasnapalavra #376.mp3'

with open(arquivo, 'rb') as audio:
    transcricao = client.audio.transcriptions.create(
        file=(arquivo, audio.read()),
        model='whisper-large-v3',
        response_format='json',
        language='pt'
    )

    format_text(transcricao)
```

```python
import base64
from groq import Groq

client = Groq()

# Função para converter imagem local em base64
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

image_path = 'image/image.jpg'
base64_image = encode_image(image_path)

completion = client.chat.completions.create(
   model='meta-llama/llama-4-scout-17b-16e-instruct', 
    messages=[{
        "role": "user",
        "content": [
            {"type": "text", "text": "O que há nessa imagem?"},
            {
                "type": "image_url",
                "image_url": {
                    "url": f"data:image/jpeg;base64,{base64_image}"
                }
            }
        ]
    }],
    temperature=1,
    max_completion_tokens=1024,
    top_p=1,
    stream=False
)

print(completion.choices[0].message.content)
```
