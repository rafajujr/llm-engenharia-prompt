```python
import os

api_key = os.environ.get("GEMINI_API_KEY")
```

```python
import google.generativeai as genai
genai.configure(api_key=api_key)
```

```python
model = genai.GenerativeModel('gemini-1.5-flat')
response = model.generate_content('Qual a dica você me recomentda para ter uma vida boa?')
```

```python
from google import genai

client = genai.Client(api_key=api_key)

response = client.models.generate_content(
    model="gemini-3-flash-preview", contents="Me diga uma dica para uma vida feliz"
)

import textwrap

text = response.text
text_formatted = textwrap.fill(text, width=100)
print(text_formatted)
```

### Stream Gemini


```python
from google import genai

client = genai.Client(api_key=api_key)
stream = client.models.generate_content(
    model="gemini-2.5-flash", contents="Qual o motivo de ser feliz?", 
)

complete_reponse = ""

for chunk in stream:
    print(chunk.text, end="", flush=True)
    complete_reponse += chunk.text
```
