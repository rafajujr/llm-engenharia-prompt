```python
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
import os
api_key = os.getenv("COHERE_API_KEY")
```

```python
import cohere
client = cohere.Client(api_key=api_key)
```

```python
model = 'command-a-03-2025'
response = client.chat(
    model=model,
    message='Ola quem é você?',
)
print(response.text)
```

```python
stream = client.chat_stream(
    model=model,
    message='Ola quem é você?',
)
for chunck in stream:
    if chunck.event_type == "text-generation":
        print(chunck.text, end='', flush=True)
```

```python
historic = [
    { "role": "user", "text": "Ola eu me chamo Rafael"},
    {"role": "CHATBOT", "text": "Ola Rafael, Como posso lhe ajudar, sou o seu assitente virtual"}
]

stream = client.chat_stream(
    model=model,
    message='Qual é o seu nome?',
    chat_history=historic
)
for chunck in stream:
    if chunck.event_type == "text-generation":
        print(chunck.text, end='', flush=True)
```
